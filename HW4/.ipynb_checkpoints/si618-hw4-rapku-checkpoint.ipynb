```python
import re
import pyspark
from pyspark import SparkContext
sc = SparkContext.getOrCreate()

totc = ["It was the best of times",
    "it was the worst of times",
    "it was the age of wisdom",
    "it was the age of foolishness",
    "it was the epoch of belief",
    "it was the epoch of incredulity",
    "it was the season of Light",
    "it was the season of Darkness",
    "it was the spring of hope",
    "it was the winter of despair",
    "we had everything before us",
    "we had nothing before us",
    "we were all going direct to Heaven",
    "we were all going direct the other way"]

WORD_RE = re.compile(r"\b[\w']+\b")

input_text = sc.parallelize(totc)
word_counts_sorted = input_text.flatMap(lambda line: WORD_RE.findall(line)) \
    .map(lambda word: (word, 1)) \
    .reduceByKey(lambda accumulator, value: accumulator + value) \
    .sortBy(lambda x: x[1], ascending = False)

top10_sorted = word_counts_sorted.take(10)
top10_sorted
```

## Bigram Analysis
Bigrams are pairs of continguous words.  For example, consider the text "The quick brown fox".  The bigrams in that text are: (The, quick), (quick, brown), (brown,fox).

You should be able to use the above code as a starting point, with the main difference being that you will be extracting pairs of words rather than single words.

1. Normalize the text by converting it to lowercase.  
1. Extract all bigrams from the text.  Don’t try to get fancy: just take all the pairs of words from each line of your text file
1. Perform a mapping to yield a count of one for each bigram (e.g. (("the", "quick"), 1)). 
1. Reduce this list of bigrams with count of one to counts of bigrams (e.g. (("the", "quick"), 312)).
1. Sort to determine the most frequent and total number of bigrams. 
1. Report the total number of bigrams, the top 20 most common bigrams and the minimum number of bigrams required to add up to 10% of all bigrams.

### Steps 2 and 3: Normalize text and extract bigrams
Note: there are many ways to accomplish this.  We recommend you create
a function that both creates bigrams and normalizes text.  The following

```python
def normalize_and_extract_bigrams(line):
    towords = line.lower()
    towords = re.findall('\w+', towords)
    words = [(towords[x], towords[x+1]) for x in range(len(towords)-1)]
    return words
```

```python
bigrams1 = input_text.flatMap(normalize_and_extract_bigrams)
bigrams1.take(20)
```

### Step 4:  Map the bigrams to key-value pairs where the key is the bigram and the value is 1

```python
bigrams2 = bigrams1.map(lambda x: (x, 1)) # Replace the ... with the appropriate code
```

### Step 5: Reduce the (word,1) key-value pairs to give you counts of each bigram

```python
bigrams3 = bigrams2.reduceByKey(lambda x, y: x+y)
bigrams3.take(5)
# Replace ... with your code
```

### Step 6: Sort the resulting RDD by value in descending order

```python
bigrams_sorted = bigrams3.sortBy(lambda x: x[1], ascending=False)
bigrams_sorted.take(5)
```

### Above and Beyond
I believe my work below merits extra credit as it attempts to explore common bigrams used in both 'The Count of Monte Cristo', and 'The Three Musketeers'. The first part explores the number of unique bigrams shared by the 2 books (33,049 bigrams), using RDD intersection to determine this.

The second part focuses on finding the number of shared total bigrams for the 2 books, exploring which combination of words led to the highest frequencies. To do this, I used the 99.9th percentile of the shared total bigrams as reference, since most word combinations will be 0. I then laid the frequency of each bigram falling under this filter in a heatmap, indicating that bigrams with the second word as 'the' had the highest common frequency between the two books.

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
musketeers = sc.textFile('Count-Musk.txt')
musketeers1 = musketeers.flatMap(normalize_and_extract_bigrams)
musketeers2 = musketeers1.map(lambda x: (x, 1))
musketeers3 = musketeers2.reduceByKey(lambda x, y: x+y)
```

```python
commonbigrams = bigrams1.intersection(musketeers1)
commonbigrams.take(5)
```

```python
print('Between the Count of Monte Cristo and The Three Musketeers, \
there are {} distinctly unique bigrams'.format(commonbigrams.count()))
```

```python
print('Unique bigrams in Count of Monte Cristo: {}\n\
Unique bigrams in The Three Musketeers: {}'.format(bigrams1.distinct().count(), musketeers1.distinct().count()))
```

```python
mixed = musketeers3.join(bigrams3)
sharedbigrams = {bigram[0]: min(bigram[1][0], bigram[1][1]) for bigram in mixed.collect()}
```

```python
sharedbigrams
```

```python
sharedbigramdf = pd.DataFrame({'Bigram':list(sharedbigrams.keys()), 'Count':list(sharedbigrams.values())})
sharedbigramdf['1st_word'] = sharedbigramdf.Bigram.map(lambda x: x[0])
sharedbigramdf['2nd_word'] = sharedbigramdf.Bigram.map(lambda x: x[1])
```

```python
filtered = sharedbigramdf[sharedbigramdf.Count >= 212]
sharedbigramdf.Count.quantile(0.999)
```

```python
matrix =filtered.pivot(index='1st_word', columns='2nd_word', values='Count').fillna(0)
matrix
```

```python
sns.heatmap(matrix)
```
