## The analyses below are focused on the businesses in United State.

```python
us_states_list = ["SC","AZ","MN","NV","WI","CA","NC","IL","AL","NM","PA","TX","FL","AK"]
```

### Step 2: Filter "df" to include only businesses in the United States.

```python
US = df.filter(df['state'].isin(us_states_list))
# change this line of code
```

### Step 3: Show the U.S. cities, and their number of businesses in descending order.

```python
US_count = US.groupby('city').count().sort('count', ascending=False)
# change this line of code
US_count.show()
```

### Step 4: Show the frequency distribution of stars in the U.S., ordered by stars from 1.0 to 5.0.

```python
US_stars_distribution = US.groupby('stars').count().sort('stars') # change this line of code
US_stars_distribution.show()
```

### Step 5a: Find the list of U.S. cities that have at least(>=) 50 businesses.

```python
cities = US_count.filter(US_count['count'] >= 50)
cities.show(1200)
# change this line of code; use multiple lines of code if needed
```

### Step 5b: Filter "US" using the list of cities from 6a, to restrict our attention to cities with at least 50 businesses reviewed.

```python
US = US.join(cities, on='city', how='inner').drop('count')
US.show()
# change this line of code
```

### Step 6a: Compute the average stars for each city.

The Spark DataFrame should look similar to this:

| --- | --- |
| San Diego | 3.5 |

```python
ratings = US.groupby('city').avg('stars')
ratings.show()
# change this line of code
```

### Step 6b: What are the top 5 cities with the highest average stars?

```python
ratings.sort('avg(stars)', ascending=False).show(5)
# change this line of code
```

### Step 7a: Define a city's "weekend friendliness" as the percentage of businesses open on Saturday and Sunday. Create a Spark DataFrame "weekend_friendliness" that gives a "weekend friendliness" for each city. 

#### Hint: use .isNotNull() in .filter()

The Spark DataFrame should look similar to this:

| city | weekend friendliness |
| --- | --- |
| San Diego | 0.5 |
| Pittsburgh | 0.5 |
| ... | ... |


```python
from pyspark.sql import functions as F
from pyspark.sql.functions import array_contains,col
#         # open on Saturday/Sunday | # total count | proportion
# city1
# city2
# change this line of code
null = []
weekend_friendliness = US.select('city', 'Hours', 'business_id', F.when((US['Hours']['Sunday'].isNotNull()) &\
                                                         (US['Hours']['Saturday'].isNotNull()), 1).otherwise(0).alias('weekend'))

weekend_friendliness = weekend_friendliness.filter(((US['Hours']['Saturday'].isNull())&\
                                             (US['Hours']['Sunday'].isNull())&\
                                             (US['Hours']['Monday'].isNull())&\
                                             (US['Hours']['Tuesday'].isNull())&\
                                             (US['Hours']['Wednesday'].isNull())&\
                                             (US['Hours']['Thursday'].isNull())&\
                                             (US['Hours']['Friday'].isNull())) == False)
weekend_friendliness = weekend_friendliness.groupby('city')\
.agg((F.sum('weekend')/F.count('weekend')).alias('weekend_friendliness')).select('city', 'weekend_friendliness')
weekend_friendliness.show()
```

### Step 7b: What are the top 5 weekend friendly cities?

```python
weekend_friendliness.orderBy('weekend_friendliness', ascending=False).show(5)
```

### Step 8a: Define a city's "nightlife hotness" as the percentage of businesses whose categories contains “Nightlife”. Create a Spark DataFrame "nightlife_hotness" that gives a "nightlife_hotness" for each city.

#### Hint1: use array_contains() to check if an element exists in a JSON array.
#### Hint2: use F.when() to make an binary variable to indicate if a business's categories contains "Nightlife".

The Spark DataFrame should look similar to this:

| city | nightlife_hotness |
| --- | --- |
| San Diego | 0.5 |
| Pittsburgh | 0.5 |
| ... | ... |


```python
from pyspark.sql import functions as F
from pyspark.sql.functions import array_contains,col


nightlife_hotness = \
US.select(US['city'], F.when(array_contains(US['categories'], 'Nightlife'), 1).otherwise(0).alias('nightyes'))\
.groupby('city').agg((F.sum('nightyes')/F.count('nightyes')).alias('nightlife_hotness'))
nightlife_hotness.show(5)
# change this line of code
```

### Step 8b: What are the top 5 cities best for nightlife?

```python
nightlife_hotness.sort('nightlife_hotness', ascending=False).show(5) # change this line of code
```

### Step 9: Join the 3 DataFrames "ratings", "weekend_friendliness", and "nightlife_hotness". Convert the joined table into Pandas DataFrame.

| San Diego | 0.5 | 0.5 | 3.5 |

```python
combine = ratings.join(weekend_friendliness, on='city', how='outer').join(nightlife_hotness, on='city', how='outer')
joined = combine.toPandas()
# change this line of code
joined.head(500)
```

##### Additional Question 1
Schema indicates ambiance, can it be separated to see if there is any curious trends on which tags are common? Does it relate to noise level?

```python
def newamb(amb):
    return F.when(US['attributes.Ambience.{}'.format(amb)] == True, 1).otherwise(0).alias(amb)

noiselevel = F.when(US['attributes']['Noise Level'] == 'quiet', 1)\
            .when(US['attributes']['Noise Level'] == 'average', 2)\
            .when(US['attributes']['Noise Level'] == 'loud', 3)\
            .when(US['attributes']['Noise Level'] == 'very_loud', 4).otherwise(0).alias('noise')

new = US.select('city', 'name', newamb('casual'), newamb('classy'), newamb('divey'),\
                newamb('hipster'), newamb('intimate'), newamb('romantic'), newamb('touristy'),\
                newamb('trendy'), newamb('upscale'), noiselevel)
new.show()
```

```python
grp = new.groupby(['city', 'noise']).sum().orderBy(['city', 'noise']).drop('sum(noise)')

for x in grp.columns:
    if x == 'city' or x == 'noise':
        continue
    grp = grp.withColumnRenamed(x, x[4:-1])
grp.show(300)
```

##### Additional question 2:
Las Vegas has the most entries from the dataset. Since the data provides latitude and longitude, can I visualize where the most noisy establishments are in Las Vegas? The quiet ones?

```python
noiseloc = US.select('name', 'city', 'latitude', 'longitude', noiselevel).filter(US.city == 'Las Vegas')
noiseloc = noiseloc.filter(noiseloc.noise != 0).toPandas()
noiseloc
```

```python
import folium
vegasmap = folium.Map(location=[36.12, -115.15], zoom_start=11)
colormap = {1:'green', 2:'yellow', 3:'orange', 4:'red'}
for x, y in noiseloc[noiseloc.noise != 2].iterrows():
    folium.CircleMarker(location=[y['latitude'], y['longitude']], color = colormap[y['noise']], radius=2).add_to(vegasmap)
```

```python
vegasmap
```

Map provides clear indication on concentration of noise on the strip. Noise level shifts primarily to quiet by immediately leaving the Vegas strip, as seen in Chinatown and Flamingo Heights.
